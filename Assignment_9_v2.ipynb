```python
import json
import csv
import pickle
```

```python
data_csv_file = open("data.csv","rt")
csv_read = csv.reader(data_csv_file)
csv_read_ls = []
for row in csv_read:
    csv_read_ls.append(row)
data_csv_file.close()
```

```python
title = csv_read_ls[0]
csv_read_ls.remove(title)
#csv_read_ls
```

```python
for i in range(len(csv_read_ls)):
    csv_read_ls[i] = csv_read_ls[i][0].split("\t")
#csv_read_ls
```

```python
for entry in csv_read_ls:
    while len(entry) < 7:
        entry.append(None)
#csv_read_ls        
```

```python
title = title[0].split('\t')
title.remove('')
#title
```

```python
index = list(range(len(csv_read_ls)))
```

```python
dict_csv = {}
for i in index:
    temp_csv_dict = {}
    for j in range(6):
        temp_csv_dict[title[(j)]] = csv_read_ls[i][(j+1)]
    dict_csv[i] = temp_csv_dict
print(dict_csv[9])
#dict_csv
```

```python
data_json_file = open("data.json","rt")
json_read = json.loads(data_json_file.read())
data_json_file.close()
#json_read
```

```python
#name_vec = []
#for key in json_read[title[0]]:
    #name_vec.append(json_read[title[0]][key])
#name_vec
```

```python
dict_labels = {}
for j in range(6):
    temp_dict_labels = {}
    temp_dict_labels_vec = []
    for key in json_read[title[j]]:
        temp_dict_labels_vec.append(json_read[title[j]][key])
    dict_labels[title[j]] = temp_dict_labels_vec
#dict_labels
   
```

```python
dict_json = {}
for i in range(20,40):
    temp_json_dict = {}
    for j in range(6):
        temp_json_dict[title[j]] = dict_labels[title[j]][(i-20)]
    dict_json[i] = temp_json_dict
print(dict_json[25])
#dict_json
```

```python
with open('data.pkl','rb') as f:
    data_pkl = pickle.load(f)
    
#data_pkl
```

```python
dict_labels_2 = {}
for j in range(6):
    temp_dict_labels_2 = {}
    temp_dict_labels_vec_2 = []
    for key in data_pkl[title[j]]:
        temp_dict_labels_vec_2.append(data_pkl[title[j]][key])
    dict_labels_2[title[j]] = temp_dict_labels_vec_2
    
```

```python
dict_pkl = {}
for i in range(40,100):
    temp_pkl_dict = {}
    for j in range(6):
        temp_pkl_dict[title[j]] = dict_labels_2[title[j]][(i-40)]
    dict_pkl[i] = temp_pkl_dict
print(dict_pkl[56])
#dict_pkl
```

```python
dict_fin = {}
dict_fin.update(dict_csv)
dict_fin.update(dict_json)
dict_fin.update(dict_pkl)
dict_fin
```

```python
countries = []
for key in dict_fin:
    if dict_fin[key]['Country'] != None:
        if dict_fin[key]['Country'] not in countries:
            countries.append(dict_fin[key]['Country'])
            
# List of Unique Countries
countries
```

```python
email = []
for key in dict_fin:
    if dict_fin[key]['Email'] != None:
        if dict_fin[key]['Email'] not in countries:
            email.append(dict_fin[key]['Email'])
#List of unique emails
email
```

```python
first_names = []
for key in dict_fin:
    if "p.o. box" not in (dict_fin[key]['Address']).lower():
        first_names.append(((dict_fin[key]['Name']).split(" "))[0])
first_names     
```

```python
sort_country = []
for key in dict_fin:
    if dict_fin[key]["Country"] != None:
        sort_country.append((dict_fin[key]["Name"],dict_fin[key]["Country"]))
sort_country.sort(key = lambda x: x[1])
first_5_names_country = []
for i in range(5):
    first_5_names_country.append(sort_country[i][0])
first_5_names_country
```

```python
sort_phone = []
for key in dict_fin:
    if dict_fin[key]["Phone"] != None:
        sort_phone.append((dict_fin[key]["Name"],dict_fin[key]["Phone"]))
sort_phone.sort(key = lambda x: x[1])
sort_phone
first_5_names_phone = []
for i in range(5):
    first_5_names_phone.append(sort_phone[i][0])
first_5_names_phone
```

```python
countries = [countries]
countries_csv_write = open('question_1.csv',"wt")
countries_write = csv.writer(countries_csv_write)
countries_write.writerows(countries)
countries_csv_write.close()
```

```python
email = [email]
email_csv_write = open('question_2.csv',"wt")
email_write = csv.writer(email_csv_write)
email_write.writerows(email)
email_csv_write.close()
```

```python
first_names = [first_names]
first_names_csv_write = open('question_3.csv',"wt")
first_names_write = csv.writer(first_names_csv_write)
first_names_write.writerows(first_names)
first_names_csv_write.close()
```
